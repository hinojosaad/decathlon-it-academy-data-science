```python
recomendations["year"]=recomendations['time_reco'].astype(str).str[:4]
recomendations["month"]=recomendations['time_reco'].astype(str).str[8:10]

recomendations["year"]= recomendations["year"].astype(int)
recomendations["month"]=recomendations["month"].astype(int)
recomendations = recomendations.drop(columns=["time_reco"])
recomendations
```

```python
products = products.rename(columns = {'modeild' : 'id_product'})
```

```python
recom_prod = recomendations.merge(products.drop_duplicates('id_product'), on = 'id_product', how= 'left')
recom_prod.head()
```
