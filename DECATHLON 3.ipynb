```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np

pd.set_option('display.max_columns', 500)
```

```python
recomedations_final = pd.read_csv('recom_prod_final.csv')
recomedations_final.head()
```

```python
recomedations_final['precio_dif'] = recomedations_final['precio_dif'].map({True: 1, False :0})
recomedations_final.head()
```

```python
recomedations_final.columns
```

```python
numerical_cols = ['product_bought', 'year', 'month', 'Review-AverageRating', 'DisplayPrice-Amount',
       'ListPrice-Amount', 'precio_dif',  'age_sender', 'age_receiver' ]
```

```python
recomedations_final[numerical_cols].describe().round(2)
```

```python
plt.figure(figsize=(15,10))
colormap = plt.cm.YlGnBu

plt.title('Pearson Correlation of features in Recommendations Data', y=1.05, size = 15)
sns.heatmap(recomedations_final[numerical_cols].corr(), linewidths=0.1,vmax=1.0, square = True, cmap = colormap, 
           linecolor = 'white', annot = True)
plt.show()
```

```python
recomedations_final.head()
```

```python
recomedations_bought = recomedations_final[recomedations_final['product_bought'] == 1]
recomedations_bought.head()
```

```python
top_10 = recomedations_bought.productnature.value_counts().to_frame()[:10]
top_10
```

```python
# reviews in prod sold 
reviews_prod_sold = recomedations_bought.groupby(['productnature'])['Review-AverageRating'].mean().to_frame().round(2)
reviews_prod_sold
```

```python
# average reviews for top 10 products 
top10_reviews = top_10.merge(reviews_prod_sold, left_index=True, right_index=True, how = 'left')
top10_reviews
```

```python
cmap = plt.get_cmap('cool')
clrs = cmap(np.linspace(1, 0, 20)) 

plt.figure(figsize = (12,8))
top10_reviews['productnature'].plot(kind = 'bar', color = clrs)
plt.title('Top 10 sold products by Product Type', fontsize = 15)
plt.xticks(rotation= 75)

plt.show()
```

```python
# average review for top 10 sold products 

cmap = plt.get_cmap('cool')
clrs = cmap(np.linspace(0, 1, 20)) 

plt.figure(figsize = (12,8))
top10_reviews['Review-AverageRating'].plot(kind = 'bar', color = clrs)
plt.title('Average Reviews for TOP 10 selling products', fontsize = 15)
plt.xticks(rotation= 75)

plt.show()
```
